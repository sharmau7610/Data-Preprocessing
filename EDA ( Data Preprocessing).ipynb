```python
def fill_numeric(data):
    for i in data.columns:
        data[i]=data[i].fillna(data[i].median())
    print(data.isnull().sum())
    return data
```

```python
def fill_categoric(data):
    for i in data.columns:
        data[i]=data[i].fillna(method='ffill')
    print(data.isnull().sum())
    return data
```

```python
#4. problem outliers
from scipy.stats  import zscore

    
```

```python
def zscore_calci(data):
    for i in data.columns:
        data[i]=np.abs(zscore(data[i]))
    x=np.where(data>3)
    data.drop(data.index[x[:1]],inplace=True)
    return data

    
```

```python
zscore_data=zscore_calci(filled_numeric)
zscore_data
```
